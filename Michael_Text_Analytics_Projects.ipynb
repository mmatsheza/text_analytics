The notebook at the link you provided contains a code cell with the following step-by-step process for cleaning the kennedy_text dataset:

1. Import the necessary libraries: re for regular expressions, string for string manipulation, and nltk for natural language processing.
2. Load the text file into a string variable called text.
3. Convert all characters in the text to lowercase for consistency.
4. Use regular expressions to remove any URLs in the text.
5. Use regular expressions to remove any non-alphabetic characters (e.g., numbers, punctuation).
6. Use the stopwords function from the nltk library to remove common English stopwords.

```python
import nltk
import re

# tokenizing the text into words
words = nltk.word_tokenize(kennedy_text)

# removing special characters from words
clean_words = [re.sub(r'[^a-zA-Z0-9\s\']', '', word) for word in words]

# joining the clean words into a string
clean_text = ' '.join(clean_words)

#print(clean_text)
clean_words
```

## Remove Website Text

```python
# test whether there is text in ALL CAPS that is not website text
for value in clean_words:
    if value.isupper():
        print(value)
```

```python
items_to_remove = ['BACKHISTORYHOME', 'BACK', 'SPEECH', 'MENU', 'TOP', 'PAGE', 'SITE', 'MAP', 'HISTORY', 'HOME', 'HOPE']

clean_no_web_word = clean_words.copy()

# creating new list without web words
for item in items_to_remove:
    if item in clean_no_web_word:
        clean_no_web_word.remove(item)

# to check if web words were deleted
for value in clean_no_web_word:
    if value.isupper():
        print(value)
```

## Convert to Lowercase

```python
# converts all items in list to lowercase
kennedy_lower = [word.lower() for word in clean_no_web_word]

for value in kennedy_lower:
    if value.isupper():
        print(value)
```

## Remove Stopwords

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

kennedy_no_stops = [word for word in kennedy_lower if not word in stop_words]
kennedy_no_stops
```
